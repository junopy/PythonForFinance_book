```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1차원 자료

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
```

```python
x= range(len(y))
plt.plot(x, y)
```

```python
plt.plot(y.cumsum())
```

### 다양한 스타일의 추가

```python
plt.plot(y.cumsum())
plt.grid(True) # grid 추가
plt.axis('tight') # 축 간격을 더 조밀하게 조정
```

#### plt.axis 옵션
인수가 없는 경우 : 현재 축의 한계 값을 반환한다   
off : 축 선과 라벨을 없앤다.  
equal : 가로축과 세로축의 스케일을 같게 한다.  
scaled : 가로축과 세로축의 스케일을 같아지도록 크기 자체를 조정한다.  
tight : 모든 자료가 보이도록 축 범위를 조밀하게 조정한다.  
image : 모든 자료가 보이도록 축 범위를 자료의 극한값으로 조정한다.  
[xmin, xmax, ymin, ymax] : 축 범위를 주어진 리스트 값으로 명시적으로 조정한다.  

```python
# plt.xlim, ylim 사용
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) -1, np.max(y.cumsum()) +1 )
```

#### 라벨 추가

```python
plt.figure(figsize=(7,4)) #figsize 를 통해 그림의 크기 (폭, 높이)로 설정
plt.plot(y.cumsum(), 'b', lw=1.5)
plt.plot(y.cumsum(), 'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

#### 표준 색깔 약자
b : 파란색  / r : 빨강색         / m : 자홍색(마젠타)  / k : 검정색  
g : 녹색    / c : 청록색 (시안)  / y : 노랑색          / w : 흰색



## 2차원 자료

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

#### 추가적인 주석 기호 

```python
plt.figure(figsize=(7,4))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

#### 크기가 전혀 다른 여러 개의 자료를 그릴 때는 y축이 하나 더 필요할 것이다.

```python
y[:,1] = y[:,1]*100
plt.figure(figsize=(7,4))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
# y축을 하나 더 상용하는 경우
fig, ax1 = plt.subplots()
  # 첫 번째 자료를 첫번째(왼쪽) 축에 대응하여 그린다.
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2 = ax1.twinx()
  # 두 번째 자료를 두번째(오른쪽) 축에 대응하여 그린다.
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
# 두 서브플롯을 구분하는 경우
plt.figure(figsize=(7,5))
plt.subplot(211)
# subplot 2*1에서 1번째의 plot
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 2nd')
```

#### 서로 다른 플롯 형식을 선택해야 하는 경우

```python
plt.figure(figsize=(9,4))
plt.subplot(121)
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('1st Data Set')
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, color='g', label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.title('2nd Data Set')
```

## 기타 플롯 유형

### 스캐터플롯

```python
y = np.random.standard_normal((1000,2))
plt.figure(figsize=(7,5))
plt.plot(y[:,0], y[:,1], 'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

#### 세 번째 차원을 가지는 스캐터 플롯

```python
c = np.random.randint(0,10, len(y))
```

```python
plt.figure(figsize=(7,5))
plt.scatter(y[:,0], y[:,1], c=c, marker='o')
plt.colorbar()
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

### 히스토그램
금융자산 수익률을 나타내는 데에 사용된다.

```python
plt.figure(figsize=(7,4))
plt.hist(y, label=['1st', '2nd'], bins=25)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

```python
# 두 자료를 히스토그램 상에 쌓아 올려서
plt.figure(figsize=(7,4))
plt.hist(y, label=['1st', '2nd'], color=['b', 'g'], stacked=True, bins=20)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

### 박스플롯

```python
fig, ax = plt.subplots(figsize=(7,4))
plt.boxplot(y)
plt.grid(True)
plt.setp(ax, xticklabels=['1st', '2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```

### 적분 구간 나타내기

```python
from matplotlib.patches import Polygon
def func(x):
    return 0.5 * np.exp(x) + 1
```

```python
a, b = 0.5, 1.5 # 적분구간
x = np.linspace(0,2)
y = func(x)
```

```python
fig , ax = plt.subplots(figsize=(7,5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# 적분값이 함수 상한과 하한 사이의 면적과 같다는 것을 보인다.
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix,Iy)) + [(b,0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

plt.text(0.5 * (a+b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
         horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
```
