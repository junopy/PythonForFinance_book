## 변동성 곡면
금융 분야에서는 3차원 플롯을 사용할 분야가 그다지 많지 않다. 그렇지만 변동성곡면은 일련의 만기와 행사가 조합에 대해 내재 변동성을 나타낸 것으로 3차원 플롯이 필요하다.  
~ 행사가는 50에서 150사이  
~ 만기는 0.5년 에서 2.5년 사이

```python
import numpy as np
```

```python
# 위의 조건에서 2차원 좌표 생성
strike = np.linspace(50,150,24)
ttm = np.linspace(0.5,2.5, 24)
strike, ttm = np.meshgrid(strike, ttm)
```

```python
strike[:2]
```

```python
#가짜 내재 변동성 생성
iv = (strike -100)**2 / (100*strike) / ttm
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                      cmap=plt.cm.coolwarm, linewidth=0.5, antialiased=True)
# rstride : 배열의 행 간격 c : 배열의 열 간격 color : 곡면 색깔 cmap 곡면에 사용될 칼라맵
# facecolors 개별 곡면 패치의 색 shade : 그림자 표시 여부
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')

fig. colorbar(surf, shrink=0.5, aspect=5)
```

```python
# 선 스타일의 변경
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30,60)
ax.scatter(strike, ttm, iv, zdir='z', s=25, c='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
```
